```python
class cirQueue:
    def __init__(self):
        self.front=-1
        self.rear=-1
        self.size=7
        self.Queue=[None for i in range(self.size)]
    def main(self):
        item=res=ch=int()
        while True:
            print ("\n------------------ Circular Queue Menu---------------\n")
            print ("\n1->Insert\n2->Delete\n3->Display\n4->Exit")
            try:
                ch=int(input('\nEnter choice...'))
            except NameError as e:
                print ('Sorry! wrong Input...')
                self.main()
            if ch == 1:
                item=int(input("Enter ITEM for insertion"))
                res=self.insert_in_queue(item)
                if res == -1:
                    print ("\nOVERFLOW")
                else:
                    print ("Your Circular Queue is: \n")
                    self.Display()
            elif ch == 2:
                item=self.del_in_queue()
                if self.front != -1 and self.rear != -1:
                    print ("\nElement delete is: ",item)
                    self.Display()
                else:
                    print ("\nUNDERFLOW")
            elif ch == 3:
                self.Display()
            elif ch == 4:
                break
            else:
                print ("\n INVALID CHOICE!")
    def insert_in_queue(self,ele):
        if((self.front == 0 and self.rear == self.size-1) or (self.front == self.rear+1)):
            return -1
        elif (self.rear == -1):
            self.front=self.rear=0
        elif (self.rear == self.size-1):
            self.rear=0
        else:
            self.rear+=1
        self.Queue[self.rear]=ele
        return 0
    def Display(self):
        print ("\nCircular Queue is:\n (Front shown as ->> and Rear as <<- and Free Spase as -)\n")
        if self.front == -1:
            return
        if self.rear >= self.front:
            for i in range(self.front):
                print ("*",end=" ")
            print ("->>",end=" ")
            for i in range(self.front,self.rear):
                print (self.Queue[i],'<-',end=" ")
            print (self.Queue[self.rear],'<<-\n')
        else:
            for i in range(0,self.rear+1):
                print (self.Queue[i],'<<-',end=" ")
            for i in range(self.rear+1,self.front):
                print ('-','->>',end=" ")
            for i in range(self.front,self.size):
                print (self.Queue[i],'<-',end=" ")
            print ("\n...... wrap around...")
    def del_in_queue(self):
        if self.front == -1:
            return -1
        else:
            ret = self.Queue[self.front]
            if self.front == self.rear:
                self.front=self.rear=-1
            elif self.front == self.size-1:
                self.front=0 #wrap round
            else:
                self.front+=1
        return ret
#-----------------------------------------------------------------------------------------------
obj=cirQueue()
obj.main()
            
            
```
